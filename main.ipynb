```python
# i is len of df
i = len(df.columns) - 1
fig, axs = plt.subplots(nrows=i + 1, ncols=1, figsize=(8, 4*i))

for j in range(i):
    j += 1
    name = "Hz" + f"{j}"
    axs[j].scatter(df['N'], df[name], label=name, s=3)
    axs[j].set_xlabel('Nodes')
    axs[j].set_ylabel('Frequency (Hz)')
    axs[j].set_title(f'Nodes vs Frequency, dataset {j + 1}')
    axs[j].legend()

# combined plot
axs[0].set_xlabel('Nodes')
axs[0].set_ylabel('Frequency (Hz)')
axs[0].set_title(f'Nodes vs Frequency, combined')
for j in range(i):
    axs[0].scatter(df['N'], df["Hz" + f"{j + 1}"], label="Hz" + f"{j + 1}", marker='o')
axs[0].legend()


plt.tight_layout()
plt.show()
```
